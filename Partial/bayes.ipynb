```python
# Punctul 1

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

observations = ['s', 'b', 'b', 'b', 's', 'b', 's', 'b', 'b', 'b']

num_heads = observations.count('s')
num_tails = observations.count('b')

alpha_prior = 1
beta_prior = 1

alpha_posterior = alpha_prior + num_heads
beta_posterior = beta_prior + num_tails

theta = np.linspace(0, 1, 500)
posterior_distribution = beta.pdf(theta, alpha_posterior, beta_posterior)

plt.figure(figsize=(8, 5))
plt.plot(theta, posterior_distribution, label=f'Beta({alpha_posterior}, {beta_posterior})', color='blue')
plt.title("Distribuția Posterioară")
plt.xlabel('Probabilitatea de a obține cap')
plt.ylabel('Densitatea de probabilitate')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Punctul 2

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

observations = ['s', 'b', 'b', 'b', 's', 'b', 's', 'b', 'b', 'b']
observations_new = ['s', 's', 'b', 's', 'b', 'b', 's', 'b', 'b', 's']

succese_1 = observations.count('s')
esecuri_1 = observations.count('b')

succese_2 = observations_new.count('s')
esecuri_2 = observations_new.count('b')

alpha_prior = 1
beta_prior = 1

# Beta(5, 7)
alpha_post_1 = alpha_prior + succese_1
beta_post_1 = beta_prior + esecuri_1

# Beta(6, 6)
# Aceasta este o distributie uniforma in comparatie cu distributia initiala care este skewed 
# Valoarea media si mediana au aceeasi valoare pentru aceasta distributie
# in comparatie cu distributia initiala (beta 5,7) unde acestea sunt diferite
alpha_post_2 = alpha_prior + succese_2
beta_post_2 = beta_prior + esecuri_2

x = np.linspace(0, 1, 1000)

y1 = beta.pdf(x, alpha_post_1, beta_post_1)
y2 = beta.pdf(x, alpha_post_2, beta_post_2)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Distribuția posteriori (Beta(5, 7))', color='blue')
plt.plot(x, y2, label='Distribuția posteriori (Beta(6, 6))', color='green')

plt.title('Distribuțiile Posteriori Beta după două experimente')
plt.xlabel('Probabilitatea de succes')
plt.ylabel('Densitatea de probabilitate')

plt.legend()
plt.grid(True)
plt.show()
```
